```python
#Top 10 Car Brands by Count
plt.figure(figsize=(12,6))
df['Brand'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Sold Car Brands')
plt.xlabel('Brand')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Number of Cars Sold by Fuel Type
fuel_counts = df['Fuel Type'].value_counts()

fuel_counts.plot(kind='barh', color='skyblue')
plt.title('Number of Cars by Fuel Type')
plt.xlabel('Count')
plt.ylabel('Fuel Type')
plt.tight_layout()
plt.show()
```

```python
#Distribution of Manual and Automatic Cars
plt.figure(figsize=(6,4))
sns.countplot(x='Transmission', data=df, hue='Transmission', palette='muted', legend=False)
plt.title('Manual vs Automatic Cars')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.show()
```

```python
# Combine Brand and Model Name into one column for clarity
df['Car'] = df['Brand'] + ' ' + df['Model Name']

# Count how many times each car appears
top_cars = df['Car'].value_counts().head(5)

# top 5 most sold cars
top_cars.plot(kind='barh', color='green')
plt.title('Top 5 Most Sold Cars')
plt.xlabel('Number of Sales')
plt.ylabel('Car')
plt.gca().invert_yaxis()  # Optional: put the most sold car at the top
plt.tight_layout()
plt.show()
```

```python
# Group by State and sum up total price
state_sales = df.groupby('State')['Price'].sum().sort_values(ascending=False).head(10)
#top 10 performing states
state_sales.plot(kind='barh', color='teal')
plt.title('Top 10 Performing States by Total Car Sales')
plt.xlabel('Total Sales Value (₹)')
plt.ylabel('State')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
#Accidental vs Non-Accidental Cars
sns.countplot(x='Accidental', data=df, hue='Accidental', palette='Set2', legend=False)
plt.title('Accidental vs Non-Accidental Cars')
plt.xlabel('Is Accidental')
plt.ylabel('Count')
plt.show()
```

```python
# Count cars per year
year_counts = df['Year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=year_counts.index, y=year_counts.values, color='skyblue')

plt.title('Number of Cars Sold per Year', fontsize=14, fontweight='bold')
plt.xlabel('Manufacture Year')
plt.ylabel('Number of Cars Sold')

# Add data labels on top of bars
for i, v in enumerate(year_counts.values):
    plt.text(i, v + 5, str(v), ha='center', fontsize=9)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
